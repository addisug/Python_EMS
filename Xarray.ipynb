# Xarray Package

- `Xarray` is an open source Python package.

- Xarray is an efficient & useful package to `read`, `edit`, and `display` spatial data.

- Xarray organizes `gridded`, `multi-dimensional`, `large datasets` and `metadata` in a human friendly format.

- Xarray allow advanced `analytic` & `visualization` methods.

- Xarray labels a multi-dimensional arrays (data) in the form of `dimensions`, `coordinates`, and `attributes`.

- Xarray depends & integrate with other python data analysis packages such as `Pandas` & `numpy`.

- Xarray allows efficent I/O operation of `NetCDF` file & parallel computing.

# Advantages of Xarray over Numpy

- numpy is a powerful but low-level in array manipulation tool.

- numpy axis's only have numbers but no names (it is easy to forget which axis is what, a common source of trivial bugs).

- numpy arrays can't carry metadata (e.g. units), and the data is unstructured (i.e. the coordinates have to be handled separately: another source of bugs).

- Like Pandas, xarray has two fundamental data structures: `DataArray` and `Dataset`:

  - `DataArray` holds a single multi-dimensional variable & it's coordinates.
  
  - `Dataset` holds multiple variables that potentially share the same coordinates.

## Calling Matplotlib functionality

```python
import matplotlib.pyplot as plt
```

```python
air_2d.plot(cmap=plt.cm.Blues)
plt.title("These colors prove North America\nhas fallen in the ocean")
plt.ylabel("latitude")
plt.xlabel("longitude")
plt.tight_layout()
plt.draw()
```

## Adjust the color scale: Robust

- Outliers often have an extreme effect on the output of the plot.

- The easy way to visualize the data without the outliers is to pass the parameter `robust=True`

- This will use the 2nd and 98th percentiles of the data to compute the color limits


```python
air_2d.plot(robust=True)
```

## Discrete Colormaps

- `levels` keyword argument can be used to generate plots with discrete colormaps.  


- For example, to make a plot with 8 discrete color intervals:

```python
air_2d.plot(levels=12)
```

## Specify the boundaries of the discrete colormap

```python
air_2d.plot(levels=[256, 276, 286, 296, 306, 316, 320])
```

## List of discrete colors through the colors argument:

```python
color_list = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
air_2d.plot(levels=[256, 276, 286, 296, 306, 316, 320], colors=color_list)
```
